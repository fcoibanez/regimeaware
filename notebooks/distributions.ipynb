```python
import pandas as pd
import numpy as np
from regimeaware.constants import DataConstants

mdl_hmm = pd.read_pickle(f"{DataConstants.WDIR.value}/data/dist/mdl_hmm.pkl")
```

```python
idx = pd.Index(["Mkt-Rf", "SMB", "HML", "RMW", "CMA", "UMD"], name="Factors")
collect_cols = {}
for state in range(mdl_hmm.n_components):
    lmb_s = pd.Series(mdl_hmm.means_[state], index=idx)
    vol_s = pd.Series(np.sqrt(np.diag(mdl_hmm.covars_[state])), index=idx)

    collect_cols[(state + 1, "lambda")] = lmb_s
    collect_cols[(state + 1, "sigma")] = vol_s

df = pd.DataFrame.from_dict(collect_cols, orient="columns")
df.style.format('{:.2%}')
```

```python
df = pd.DataFrame(mdl_hmm.transmat_)
df.index = [f"From State {i + 1}" for i in range(df.shape[0])]
df.columns = [f"To State {i + 1}" for i in range(df.shape[1])]
df.style.format('{:.2%}')
```
